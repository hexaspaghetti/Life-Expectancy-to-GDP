# Life Expectancy of Countries

### Goals
    Derive meaningful information from life expectancy data.
### Data 
    Data will be provided from codecademy and contains info about The a country, the life expectancy of country, its gdp, and the year its stats were taken.  
### Analysis
    Performing a descriptive analysis to show the relationship between life expectancy, gdp and time. 

To start off we are going to import the libraries I'll be using for the data visualization. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Then I'll extract the data into a dataframe and analyze the first few rows to show its general structure.

```python
df = pd.read_csv('all_data.csv')
print(df.head())
print(df.dtypes)
```

Here I'm just storing the country names for later use for labels or reference.

```python
countries = df.Country.unique()
country_labels = ['Chile', 'China', 'Germany', 'Mexico','United States of America', 'Zimbabwe']
```

## Average life expectancy per country (bar chart)

Below is the code we'll be using to generate a bar chart to show the average life expectancies of countries.

```python
ax = plt.subplot(1,1,1)
plt.bar(countries, df.groupby('Country')['Life expectancy at birth (years)'].mean())
plt.xlabel('Countries')
plt.ylabel('Life Expectancy At Birth (Years)')
plt.title('Average Life Expectancy Per Country')
plt.style.use('seaborn')
ax.set_ylim([40,85])
ax.set_xticks(range(len(country_labels)))
ax.set_xticklabels(country_labels, rotation=30)
plt.show()
plt.close()
```

From this chart I've noticed that Germany has the highest life expectancy out of these countries, with Zimbabwe being drastically lower. 

## Plot Average life expectancy over time per country (line plot)

Now below is the code that will be used to show the relation of life expectancy over time.

```python
x = 1
for country in country_labels:
    ax2 = plt.subplot(2,3,x)
    plt.subplots_adjust(hspace=.35, wspace=.35)
    country_data = df[df['Country']==country]
    plt.plot(country_data.Year, country_data['Life expectancy at birth (years)'], marker='.')
    plt.title(country)
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy')
    
    x +=1
plt.show()
plt.close()
```

There seems to be a clear correlation that generally over time life expectancy goes up.

## Plot life expectancy per countries gdp (scatter)

Finally this is the code written to plot the life expectancy against gdp to see  the correlation between the two.

```python
x2 = 1
for country in country_labels:
    ax3 = plt.subplot(2,3,x2)
    plt.subplots_adjust(hspace=.35, wspace=.45)
    country_data = df[df['Country']==country]
    plt.scatter(country_data.GDP, country_data['Life expectancy at birth (years)'], marker='.')
    plt.title(country)
    plt.xlabel('GDP')
    plt.ylabel('Life Expectancy')
    
    x2 +=1
plt.show()
plt.close()
```

As we can see here, there is a clear correlation between GDP and life expectancy. 

## Conclusion
So through the analysis we can see a few things:
    * There is a positive correlation of time to life expectancy.
    * There is also a positive correlation of GDP to life expectancy.
    * From this data Zimbabwe has the lowest GDP and mean life expectency.
    * From this data we also see how Germany has the highest GDP and the highest mean life expectency.
So it seems that we can predict as GDP goes up, life expectancy will follow.
